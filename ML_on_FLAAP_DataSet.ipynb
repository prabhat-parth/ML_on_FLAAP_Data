```python
flaap['Activity'].value_counts().plot(kind='bar', title='Number of Examples by Activities',color=['b','r','g','y','k','r']);
```

```python
def plot_activity(activity, df):
    data = df[flaap['Activity'] == activity][['acc_x', 'acc_y', 'acc_z']][:200]
    axis = data["acc_x"].plot(subplots=True, 
                     title=activity,color="b")
    axis = data["acc_y"].plot(subplots=True, 
                 title=activity,color="r")
    axis = data["acc_z"].plot(subplots=True, 
             title=activity,color="g")
    
    for ax in axis:
        ax.legend(loc='lower left', bbox_to_anchor=(1.0, 0.5))
    plt.show()
```

```python
for activity in np.unique(flaap['Activity']):
    plot_activity(activity, flaap)
```

```python
def plot_activity(activity, df):
    data = df[flaap['Activity'] == activity][['gyr_x',	'gyr_y',	'gyr_z']][:200]
    axis = data["gyr_x"].plot(subplots=True, 
                     title=activity,color="b")
    axis = data["gyr_y"].plot(subplots=True, 
                 title=activity,color="r")
    axis = data["gyr_z"].plot(subplots=True, 
             title=activity,color="g")
    
    for ax in axis:
        ax.legend(loc='lower left', bbox_to_anchor=(1.0, 0.5))
    plt.show()
```

```python
classifiers = [
    linear_model.LogisticRegression(max_iter=100),
    KNeighborsClassifier(3),
    RandomForestClassifier(n_estimators=100),
    AdaBoostClassifier(),
    GaussianNB()
    ]
```

```python
# Logging for Visual Comparison
log_cols=["Classifier", "Accuracy", "Log Loss"]
log = pd.DataFrame(columns=log_cols)

for clf in classifiers:
  print(clf)
  trained_model = clf.fit(X_train, y_train)
  name = clf.__class__.__name__    
  print("="*30)
  print(name)       
  print('****Results****')
  # Predict missing values' class
  imputed_values = clf.predict(X_test)
  acc = accuracy_score(imputed_values, y_test)
  print("Accuracy: {:.4%}".format(acc))
  precision = precision_score(imputed_values, y_test, average='macro')
  print('Precision: %f' % precision)
  # recall: tp / (tp + fn)
  recall = recall_score(imputed_values, y_test, average='macro')
  print('Recall: %f' % recall)
  # f1: 2 tp / (2 tp + fp + fn)
  f1 = f1_score(imputed_values, y_test, average='macro')
  print('F1 score: %f' % f1)
  print(confusion_matrix(imputed_values, y_test))
  print(classification_report(imputed_values, y_test))
  print("complete !!!")

print("="*30)
```
